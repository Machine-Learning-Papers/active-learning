# # Main functions

```python
def download_mnist(path="./data/mnist/"):
    '''
    Скачивает базу MNIST (~55мб)
    '''
    mnist = fetch_mldata('MNIST original', data_home=path)
    pickle.dump(mnist, open("./data/mnist/mnist.pickled", "w"), -1)
    

def load_mnist():
    '''
    Загружает базу MNIST из дампа и уменьшает размерность до 20 (PCA)
    '''
    mnist = pickle.load(open("./data/mnist/mnist.pickled", "r"))
    pca = decomposition.PCA(n_components=20)
    logging.info("Start PCA...")
    pca.fit(mnist['data'][1:20000,:])
    X_raw = pca.transform(mnist['data'][1:40000,:])
    logging.info("Done!")
    return (X_raw, mnist['target'][1:40000], 'mnist')


def criteria(vector, method='margin'):
    '''
    Функция ранжирования определения степени уверенности.
    @vector - вектор оценок вероятностей принадлежности объекта к классам.
    @method - способ ранжирования, реализованы 3 варианта: 
                1) margin - "зазор" между двумя наиболее вероятными классами
                2) minmax - высота пика распределения
                3) random - случайное число, для сравнения с другими
    Возвращает скаляр.
    '''
    if method == 'margin':
        res = 0
        x = np.sort(-vector)
        res =  x[0] - x[1]
    if method == 'minmax':
        res = max(vector)
    if method == 'random':
        res = np.random.uniform()
    return res


def active_learning(clf, X, y, test_x, test_y, points_per_step=500, init_size=500, label='unlabeled',
                    method='margin', verbosity=False):
    '''
    Реализует итеративный процесс обучения модели и наращивания обучающего множества.
    @clf - модель классификатора. Требуется, чтобы мог возвращать вероятности (predict_proba или 
            desicion_function, который потом заворачивается в сигмоид)
    @X, @y - пул доступных для обучения точек. На стадии инициализации из него будет случайно выбрано подмножество
            размера @init_size. На каждой итерации в обучение будут добавляться @point_per_step наиболее неуверенно 
            проклассифицированных на предыдущей итерации точек.
    @text_x, @test_y - множество, используемое для оценки качества. Используются 0\1 функция потерь.
    @method - способ ранжирования точек, функция критерия
    @label - используется для подпаси графиков и сохранения результата.
    @verbosity - выводить ли поясняющую информацию.
    
    Возвращает структуру, в которой хранятся: 
        1-2) размер обучающего множества и соотвествующее ему качество классификатора
        3) множество обучающих множеств на каждом шаге
        4) описание (label)
        5) модель классификатора
    '''
    # множество индексов элементов в @X, которые используются для обучения 
    train_ids = np.array(range(init_size))
    acc = []
    train_size = []
    train_list = []
    mean_margin = []
    # множество индексов элементов в @X, которые пул ( = не используются для обучения) 
    pool_ids = np.array(range(0, len(y)))
    
    if verbosity:
        print "Init AL with X:",  X.shape, "\n\ty:", y.shape, "\n\tclf:", clf  
    
    # пока не закончились точки в пуле
    while len(pool_ids) > points_per_step:
        # перестраиваем множества train и pool
        train_y = y[train_ids]
        train_x = X[train_ids,:]
        pool_x = np.delete(X, train_ids, 0)
        pool_ids = np.delete(np.array(range(0, len(y))), train_ids)
        # сохраняем текущее обучающее множество и его размер
        train_size.append(len(train_ids))
        train_list.append(train_ids)

        # тренируем классификатор
        model = clf
        model.fit(train_x, train_y)

        # находим распределение вероятностей для точек из пула
        ###############
        try:
            probabilities = model.predict_proba(pool_x)
        except:
            tmp_prob = model.decision_function(pool_x)
            tmp_prob = 1.0/(1.0+np.exp(-tmp_prob))
            if len(tmp_prob.shape)==1:
                probabilities = np.zeros([len(tmp_prob), 2])
                probabilities[:,0] = tmp_prob
                probabilities[:,1] = 1 - tmp_prob
            else:
                probabilities = tmp_prob
        ################
        # вычисляем для них значение функции критерия
        uncertain = map(lambda x: criteria(x, method), probabilities)

        # ранжируем индексы точек по значению критерия
        res = np.zeros([len(pool_ids), 2])
        res[:, 0] = pool_ids
        res[:, 1] = uncertain
        scores = np.array(sorted(res, key=lambda w: -w[1]))

        # проверяем качество текущей модели
        pred = model.predict(test_x)
        accuracy = np.mean(pred == test_y)
        acc.append(accuracy)
        if verbosity:
            print(len(train_ids), accuracy)
        
        mean_margin.append(mean(scores[0: points_per_step, 1]))
        # наращиваем обучающее множество наиболее неуверенными точками
        train_ids = np.concatenate([train_ids, np.array(scores[0: points_per_step, 0]).astype(int)])

        del train_x
        del train_y
        del pool_x
        del model
        gc.collect()
        
    return (train_size, acc, train_list, label, clf, mean_margin)
```

```python
# download data
logging.info('Downloading MNIST dataset (~55mb, need only one time)...')
download_mnist()
logging.info('..done!')
```

```python
# load data
data = load_mnist()
print "Data loaded", data[0].shape
TRAIN_SIZE = 10000
TEST_SIZE  = 8000
PPS = 500       # points_per_step
INIT_SIZE = 500 # init_train_size
# split data
X_train, X_test, y_train, y_test = train_test_split(data[0], data[1], test_size=0.5, random_state=2014)

X_train = X_train[:TRAIN_SIZE,:]
y_train = y_train[:TRAIN_SIZE]

X_test = X_test[:TEST_SIZE,:]
y_test = y_test[:TEST_SIZE]

print "train dim:", X_train.shape
print "test dim:", X_test.shape
```

```python
param_list = [
    (ensemble.RandomForestClassifier(n_estimators=500, n_jobs=-1, max_depth=5), 'rf.est=500.depth=5'),
    (linear_model.RidgeClassifier(),"ridge.default"),
    #(linear_model.LogisticRegression(), 'logReg.default'),
    #(lda.LDA(), 'lda.default')
]

for (clf, label) in param_list:
    logging.info('Start to process:%s'%label)
    # do AL 
    al_res = active_learning(clf, X_train, y_train, X_test, y_test, PPS, INIT_SIZE, label, method='margin', verbosity=False)
    # do random
    rnd_res = active_learning(clf, X_train, y_train, X_test, y_test, PPS, INIT_SIZE, label, method='random', verbosity=False)
    logging.info("%s done!" % label)
    # plotting
    figure(figsize=(10, 8))
    title("Dataset:" + data[2] + "\nModel:" + clf.__str__())
    plot(al_res[0], al_res[1])
    plot(rnd_res[0], rnd_res[1])
    xlabel("Train size")
    ylabel("Accuracy")
    grid()
    # save plot in ./results/mnist folder!
    savefig('./results/'+data[2]+"/"+label+".png", bbox_inches='tight')
    # dumping results
    #pickle.dump(al_res, open("./results/"+data[2]+"/"+data[2]+"."+label+".pickled", 'w'), -1)
```
